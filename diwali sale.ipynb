```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.describe()
```

PLOTTING GRAPH ON BASIS OF MALE AND FEMALE BUYERS

```python
gen=sns.countplot(x='Gender',data=df)
for bars in gen.containers:
    gen.bar_label(bars)
#This shows the total number of male and female customers
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
salesgenplot=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in salesgenplot.containers:
    salesgenplot.bar_label(bars)
```

```python
print(sales_gen)
type(sales_gen)
```

This shows that Female customers have spent 4,24,33,577 INR more than male customers

AGE

```python
order = sorted(df['Age Group'].unique())
ag=sns.countplot(data=df,x='Age Group',order=order)
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
order = sorted(df['Age Group'].unique())
ag=sns.countplot(data=df,x='Age Group',hue='Gender',order=order)
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_age
#order = sorted(df['Age Group'].unique())
salesageplot=sns.barplot(x='Age Group',y='Amount',data=sales_age)
for bars in salesageplot.containers:
    salesageplot.bar_label(bars)
```

WE CAN SEE THAT PEOPLE OF AGE GROUP 26-35 SPEND THE MOST

STATE

```python
st=df.groupby(['State'],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False)
#st.head
st
```

```python
stbar=sns.barplot(x='State',y='Orders',data=st)

sns.set(rc={'figure.figsize':(30,5)})
for bars in stbar.containers:
    stbar.bar_label(bars)
```

```python
stamt=df.groupby(['State'],as_index=False)['Amount',].sum().sort_values('Amount',ascending=False)
stamt
```

```python
stamtgen=df.groupby(['State','Gender'],as_index=False)['Amount',].sum().sort_values('Amount',ascending=False)
stambar=sns.barplot(x='State',y='Amount',data=stamtgen,hue='Gender')
for bars in stambar.containers:
    stambar.bar_label(bars)
```

by the bars plotted above we can conclude that uttar pradesh has the most number of orders followed by maharashtra and karnataka
also customers from up,maharashtra and karnataka have spent the most amount of money 

PRDUCT CATEGORY VS GENDER

```python
order = df['Product_Category'].value_counts(ascending=False).index
pcgenbar=sns.countplot(x='Product_Category',data=df,hue="Gender",order=order)
for bars in pcgenbar.containers:
    pcgenbar.bar_label(bars)
```

product vs orders

```python
prord=df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
prord
```

```python
prord=df.groupby(['Product_Category','Gender'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False)
prordbar=sns.barplot(x='Product_Category',y='Orders',hue='Gender',data=prord)
for bars in prordbar.containers:
    prordbar.bar_label(bars)
```

we can conclude that people have place the most orders on clothing and apparel followed by food

```python
prord=df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
prordbar=sns.barplot(x='Product_Category',y='Amount',hue='Gender',data=prord)
for bars in prordbar.containers:
    prordbar.bar_label(bars)
```

we can see that people have spent the most money on food followed by footwear and shoes
